```python
import pandas as pd
google = pd.read_csv("google-play-store.csv")
```

```python
google.info()
```

```python
#كولوم ريليسد دا الداتا فية ديت تايم لكن التايب بتاعه ستير فلازم يتحول لديت تايم
google["released"] = pd.to_datetime(google["released"])
```

```python
#عايزة اشوف شكل الداتا فريم علشان افهم كل كولم اى بالظبط نوع الداتا اللى فية 
google.head()
```

```python
#هجممع عدد كل الميسينج فاليوز علشان اهندلها
google.isnull().sum()
```

```python
#همسك كولوم كولم اعمل فيل نا بالميدن لانة هو اللى مش هيتأثر بالاوت لاير زى المين و هيخلى الداتا قريبة للصح 
#عملت جروب باى جينرى للرفيورز و حسبت الميدين ليها و هعمل بقيمة الميدن فيل للنل فاليوز و هكذا فى كل كولم
#استخدمنا فنكشن تراتنسفورم علشان نحافظ على الداتا لكل رو زى فكرة الويندو فنكشن هتعمل عملية على جروب كبير لكن هتحافظ على ديتيلز لكل رو 
google['reviews'] = google['reviews'].fillna(
    google.groupby('genre')['reviews'].transform('median')
)
google['ratings'] = google['ratings'].fillna(
    google.groupby('genre')['ratings'].transform('median')
)
google['score'] = google['score'].fillna(
    google.groupby('genre')['score'].transform('median')
)
```

```python
#حسبت المدين للكولوم ريليس و هعمل فيل للميسينج فاليو بيه و جايبة الميدين لكل جينيرى 
google['released'] = google['released'].fillna(
    google.groupby('genre')['released'].transform('median')
)
```

```python
google.info()
```

```python
#اشوف الرو الفاضى و اعرف هل فيه داتا ولا لاء
google[google['genre'].isnull()]
```

```python
#مليت الفليو الميسنج بان نون لان اسم الاب موجود و ممكن احتاجها فى التحليل لو هصنف اسم الاب فيما بعد
google['genre'] = google['genre'].fillna('Unknown')
google['genre_id'] = google['genre_id'].fillna(-1)
```

```python
#غيرت اسم الداتا فريم 
apps_df = google
```

```python
#عملت جروب باى جينرى بالكولوم ريت بير ديت و حسبت المين و رتبت الفليوز ديسيندنج
apps_df.groupby('genre')['ratings_per_day'].mean().sort_values(ascending=False)
```

```python
#عملت جروب باى جينرى بالكولوم ريت بير ديت و حسبت المين و رتبت الفاليوز من الصغير للكبير
apps_df.groupby('genre')['ratings_per_day'].mean().sort_values()
```

```python
#عملت جروب باى للجينرى بالكولوم ريت بير ديت و هاخد اعلى اندكس رقم الصف اما الفنكشن لوك رجعتلى بياتات الرو كاملة 
apps_df.loc[apps_df.groupby('genre')['ratings_per_day'].idxmax()]
```

```python
#عملت جروب باى جينرى  على  عمود ريتنج بيرداى و حسبت المتوسط علشان ميتاثرش بالاوت لاير و عملت سورت فاليو من الكبير للصغير علشان اعرف مين اكترالتطبيقات اعلى متوسط  تقيمات يومية
apps_df.groupby('genre')['ratings_per_day'].median().sort_values(ascending=False)
```

```python
#عايزة اعرف مين التطبيقات اللى فيها خاصية الدفع من التطبيق اول حاجه فلترة بكولوم الابرايس من الداتا فريم لو البرايس اكبر من 0 و خزنتها في الداتا فريم بدل م ترجع بولين فاليبو 
#بعد كدا عملت جروب باى جينرى بالكولوم ريتنج بير داى و حسبت الميدين و رتبت الفليوز من الكبير للصغير علشان اعرف كل جينرى بيحقق اعلى معدل تقييمات يومية
paid_apps = apps_df[apps_df['price'] > 0]
paid_apps.groupby('genre')['ratings_per_day'].median().sort_values(ascending=False)
```

```python
# حسبنا متوسط التقييمات لكل جنرى و سورت للفلاليو من الكبير للصغير علشان اعرف انهى تطبيقات بتاخد اعلى معدل تقييمات 
apps_df.groupby('genre')['ratings'].mean().sort_values(ascending= False)
```

```python
# رتبنا الريفيوز من الكبير للصغير و جبنا اكبر 5 تقييمات 
apps_df.sort_values('reviews', ascending=False).head(5)
```

```python
# بتحسب كام مرة القيمة فى عمود offers_iapو كل مرة ترو بيقى التطبيق بيدعم و الفولس لا يدعم الدفع
apps_df['offers_iap'].value_counts()
```

```python
# جمعنا عدد مرات التحميل لكل جنرى 
apps_df.groupby('genre')['min_installs'].sum()
```

```python
# عايزة اعرف اقدم واحدث تاريخ
apps_df['released'].agg(['min', 'max'])
```

```python
#بنفلتر الريفيوز الاكبر من 10000 للعمود ريتنج بير داى و فى الاخر حبنا المتوسط علشان اعرف متوسط التقيمات اليومية للتطبيقات اللى عندها مراجعات كبيرة
apps_df[apps_df['reviews'] > 10000]['ratings_per_day'].mean()
```

```python
# جروب باى على عمود ريتنج و حسبنا المتوسط و رتبنا النتائج من الكبير للصغير علشان اخد اكبر 3 تطبيقات ليها ريتنج عالى
apps_df.groupby('genre')['ratings'].mean().sort_values(ascending=False).head(3)
```

```python
#شرطين لازم يتحققوا سوا الاول التطبيقات اللى بتدعم الاعلانات و التانى التطبيقات اللى بتدعم الدفع من التطبيق 
# و حسبنا متوسط التطبيقات اللى بتعدم الرشطين ثم حولناها المتوسط لنسبة مئوية
((apps_df['ad_supported'] == True) & 
 (apps_df['offers_iap'] == True)).mean() * 100
```

```python
#عايزن نحسب اكتر سنه اتحقق فيها تحميلات للتطبيقات
apps_df['released'].dt.year.value_counts().idxmax()
```

```python
# معامل الارتباط بين الرفيوز و الريتنج و عملنا الارتباط دا بعد جروبي باى جنرى 
#   لوك فانكشن علشان ترجع بيانات الرو اللى اتحقق فيه معامل الارتباط لكل جنرى بين ريفيوز و الريتنجس  
correlation_by_genre = apps_df.groupby('genre')[['reviews', 'ratings']].corr()

correlation_by_genre.loc[ : , ('reviews', 'ratings')]
```

```python
# جربنا الريتنج فى الريفيو لكل نوع جنرى و جمعنا النتيجة و قسمناها على الرفيوز لكل جنرى و مع تجميع النتيجة علشان نعرف متوسط الويت 
weighted_avg = (
    (apps_df['ratings'] * apps_df['reviews'])
    .groupby(apps_df['genre'])
    .sum()
    /
    apps_df['reviews'].groupby(apps_df['genre']).sum()
)

weighted_avg
```

```python
# فلترنا الداتا علش\ان استبعد اللى بتساوى صفر 
# نحسب النسبة
# نجيب اعلى التطبيقات لوك ترجعلنا بيانات الرو اللى معملو فية جروب باى جنرى مع الكولوم ريشسيو و اخدنا اعلى اندكس 
#اخر حاجه عرضت التوم اب بالكوم  تايتل و نوع و الريشيو يعنى جدول مختصر 
apps_df = apps_df[apps_df['ratings'] > 0]
apps_df['ratio'] = apps_df['score'] / apps_df['ratings']
top_apps = apps_df.loc[apps_df.groupby('genre')['ratio'].idxmax()]
top_apps[['title', 'genre', 'ratio']]
```
